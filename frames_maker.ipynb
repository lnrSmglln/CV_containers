Раскадровка видео

```python
import cv2
import numpy as np
```

```python
# сохранять каждый {per} кадр из нормальных интервалов
per = 5

cam = cv2.VideoCapture("CNTRL.mp4")

intervals = np.array([[255, 296], [342, 823], [1113, 2381], [3969, 4278], [4322, 4642], [5046, 6916]])
intervals = intervals + 2

current_interval = 0
current_frame = 0
i = 0

while(True):
      
    # reading from frame 
    ret, frame = cam.read() 
  
    if ret and current_interval < intervals.shape[0]:
        if current_frame >= intervals[current_interval, 0] and current_frame <= intervals[current_interval, 1] and current_frame%per == 0:
            # if video is still left continue creating images 
            name = './frames_per5/frame' + str(i) + '.jpg'
            print ('Creating...' + name) 
    
            # writing the extracted images 
            cv2.imwrite(name, frame) 
    
            i += 1

        current_frame += 1

        if current_frame > intervals[current_interval, 1]:
            current_interval += 1
    else:
        break
```

```python
# сохранить все кадры

cam = cv2.VideoCapture("CNTRL.mp4")

current_frame = 0

while(True):
      
    # reading from frame 
    ret, frame = cam.read() 
  
    if ret:
        
        # if video is still left continue creating images 
        name = './video_frames/frame' + str(current_frame) + '.jpg'
        print ('Creating...' + name) 

        # writing the extracted images 
        cv2.imwrite(name, frame) 

        current_frame += 1
    else:
        break
```
